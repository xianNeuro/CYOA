```python
# recall mat: three subjects' recall temporal order

import xianfunc as xf
import matplotlib.pyplot as plt
import matplotlib as mpl

root = './5_recall_mat/'
outdir = root
fls = xf.grab_all(root+'data/','recall.xlsx')

for f in fls:
    subj = f.split('_rate')[0]
    print(subj)
    
    rcl = xf.read_data(root+'data/'+f,[1])['recalled_events']
    rcl = xf.split_num_ls_remain(rcl,thres=6)
    fname = f.split('rate-recall')[0]+'events.xlsx'
    tot = int(xf.read_data(root+'data/'+fname,[1])['event'][-1])

    # Configure Matplotlib to ensure text remains editable in SVG
    mpl.rcParams['svg.fonttype'] = 'none'  # Prevents conversion of text to paths
    mpl.rcParams['font.family'] = 'sans-serif'  # Use a standard, simple font
    mpl.rcParams['font.sans-serif'] = ['Arial']  # Explicitly set a common font like Arial
    mpl.rcParams['text.usetex'] = False  # Disable LaTeX to avoid any unwanted path conversion

    # Set x-tick and y-tick label font sizes
    mpl.rcParams['xtick.labelsize'] = 40  # Set x-tick label font size
    mpl.rcParams['ytick.labelsize'] = 40  # Set y-tick label font size

    
    mat = xf.transpose_mat(xf.recall_temp_mat(rcl,tot))
#     xf.plot_2dmat(mat,title=subj+'_Recall_Sequence',outpath=outdir,figsize=(8,10))
    xf.plot_2dmat(mat,title=subj+'_Recall_Sequence',outpath=outdir,figsize=(8,10),fontsize=20,fig_format='.svg')
```
